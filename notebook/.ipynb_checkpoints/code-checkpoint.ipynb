## Classification (using ML & DL)

```python
## Importing Libraries


import pandas as pd
from pandas_profiling import ProfileReport
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import numpy as np
from matplotlib import pyplot
from sklearn.metrics import accuracy_score
from random import sample
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import cohen_kappa_score
```

```python
## Reading input data 


data_1 = pd.read_csv('../data/datatraining.txt', sep=",", header=0, error_bad_lines=False)
data_2 = pd.read_csv('../data/datatest.txt', sep=",", header=0, error_bad_lines=False)
data_3 = pd.read_csv('../data/datatest2.txt', sep=",", header=0, error_bad_lines=False)
n = 1000
data_3[-n:].to_csv("../data/test_data.csv", index = False)
data_3.drop(data_3.tail(n).index,inplace=True)
```

```python
data_3.sample(5, random_state=5)
```

```python
## combining the data from all the files (except last 1000 rows from datatest2)


df1 = data_1.append([data_2, data_3]).sort_values(by = "date", ascending = True).reset_index()
df1['date'] = pd.to_datetime(df1['date'])
df1 = df1.drop("index", axis = 1)
df1.head(3)
```

```python
print ("Total Obs df1 : ", df1.shape[0])
print ("Total cols df1 : ", df1.shape[1])
print ("Columns : ", list(df1.columns))
```

```python
pd_profile = ProfileReport(df1, title='Pandas_Profiling_Report')
```

```python
## Basic EDA required for this classification ( all the insights can be obtained from this profiling.. 
## for instance like missing values, total count, class imbalances, correlation etc..)
pd_profile
```

```python
## Adding extra features from timestamp which is day of the week, ex 1 = Monday, hour of the day, ex 1,2,3...24
## Without adding these features itself we are getting good accuracy
## Splitting data into train and test

df1['day'] = df1['date'].dt.dayofweek
df1['hour'] = df1['date'].dt.hour
y = df1.Occupancy
X = df1.drop(['Occupancy', 'date'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0, random_state=2) 

# split size is zero coz already trained and check accuracy. so used all the data for training, coz we have separate 1000 test data
```

```python
# Training Random Forest model

Random_forest_model = RandomForestClassifier(random_state = 3)
Random_forest_model.fit(X_train, y_train)

print ("RF Model Training Completed...")
```

```python
## Feature importance

importance = Random_forest_model.feature_importances_
cols = X_train.columns

# feature importance with scores

print ('Feature importance with scores : \n')
for i,v in enumerate(importance):
    print(str(cols[i]) , ' : ', str(v))
    
# plotting 

print ("\n\n Plot : ")
pyplot.bar([x for x in range(len(importance))], importance)
pyplot.show()


## Even though some of the features are showing not very important, still model was giving excellent accuracy i havent removed any feature.
## Some of the variables are also cor-related ( check profiling )
```

```python
## Reading test data and adding the features created while training ( day, hour )


test_data = pd.read_csv("../data/test_data.csv")
test_data['date'] = pd.to_datetime(test_data['date'])
test_data['day'] = test_data['date'].dt.dayofweek
test_data['hour'] = test_data['date'].dt.hour
test_data_y = test_data.Occupancy
test_data_d = test_data.drop(["date","Occupancy"], axis=1)
```

```python
# Predict 

pred_y_4 = Random_forest_model.predict(test_data_d)
print(pd.DataFrame( pred_y_4 , columns = ["pred"])['pred'].value_counts() )
```

```python
print ("Accuracy Metrics of RF model : \n")


# Accuracy
accuracy = accuracy_score(test_data_y, pred_y_4)
print('Accuracy: %f' % accuracy)
# precision 
precision = precision_score(test_data_y, pred_y_4)
print('Precision: %f' % precision)
# recall
recall = recall_score(test_data_y, pred_y_4)
print('Recall: %f' % recall)
# f1 score
f1 = f1_score(test_data_y, pred_y_4)
print('F1 score: %f' % f1)
# kappa
kappa = cohen_kappa_score(test_data_y, pred_y_4)
print('Cohens kappa: %f' % kappa)
```

```python
test_data['pred_RF'] = pred_y_4
```

```python
from sklearn.svm import LinearSVC
svm = LinearSVC(random_state = 2) 
  
# training model

svm.fit(X_train, y_train) 
print ("SVM model tarining completed...")
```

```python
# Predict 

pred_svm = svm.predict(test_data_d)
print(pd.DataFrame( pred_svm , columns = ["pred"])['pred'].value_counts() )
```

```python
test_data['pred_SVM'] = pred_svm
```

```python
print ("Accuracy Metrics of SVM model : \n")


# Accuracy
accuracy = accuracy_score(test_data_y, pred_svm)
print('Accuracy: %f' % accuracy)
# precision 
precision = precision_score(test_data_y, pred_svm)
print('Precision: %f' % precision)
# recall
recall = recall_score(test_data_y, pred_svm)
print('Recall: %f' % recall)
# f1 score
f1 = f1_score(test_data_y, pred_svm)
print('F1 score: %f' % f1)
# kappa
kappa = cohen_kappa_score(test_data_y, pred_svm)
print('Cohens kappa: %f' % kappa)
```

```python
## Tarining a simple MLP model for binary classification

from tensorflow import keras


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0, random_state=2)

model = keras.Sequential()
model.add(keras.layers.Dense(100, input_dim=X_train.shape[1], activation='relu'))
model.add(keras.layers.Dense(1, activation='sigmoid'))

# compile model

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# fit model

history = model.fit(X_train, y_train, validation_data=(test_data_d, test_data_y), epochs=300, verbose=0)

```

```python
# Predict

dl_preds = history.model.predict_classes(test_data_d)
print(pd.DataFrame( dl_preds , columns = ["pred"])['pred'].value_counts() )
```

```python
test_data['pred_MLP'] = dl_preds
```

```python
print ("Accuracy Metrics of MLP model : \n")


# Accuracy
accuracy = accuracy_score(test_data_y, dl_preds)
print('Accuracy: %f' % accuracy)
# precision 
precision = precision_score(test_data_y, dl_preds)
print('Precision: %f' % precision)
# recall
recall = recall_score(test_data_y, dl_preds)
print('Recall: %f' % recall)
# f1 score
f1 = f1_score(test_data_y, dl_preds)
print('F1 score: %f' % f1)
# kappa
kappa = cohen_kappa_score(test_data_y, dl_preds)
print('Cohens kappa: %f' % kappa)
```

```python
test_data.to_csv("../data/predicted.csv", index = False)
```
